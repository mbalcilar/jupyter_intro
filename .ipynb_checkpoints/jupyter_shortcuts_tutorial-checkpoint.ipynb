## 4. Create a small dataset

In data visualization, we usually move from data to a chart.

```python
import pandas as pd

scores = pd.DataFrame({
    "student": ["A", "B", "C", "D", "E", "F"],
    "hours_studied": [1, 2, 3, 4, 5, 6],
    "quiz_score": [62, 68, 73, 78, 85, 91]
})

scores
```

## 5. Plot the data with plotnine

This uses the same grammar-of-graphics idea as ggplot2, but in Python.

```python
from plotnine import ggplot, aes, geom_point, geom_smooth, labs, theme_minimal

(
    ggplot(scores, aes(x="hours_studied", y="quiz_score"))
    + geom_point(size=3)
    + geom_smooth(method="lm", se=False)
    + labs(
        title="Study time and quiz score",
        x="Hours studied",
        y="Quiz score"
    )
    + theme_minimal()
)
```
